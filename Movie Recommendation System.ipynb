# Data Preprocessing

```python
# Check for missing data
movies.isnull().sum()
```

```python
#drop missing entries
movies = movies.dropna()
movies.isnull().sum()
```

```python
#Check for duplicate entries
movies.duplicated().sum()
```

```python
movies.reset_index(inplace=True)
```

### Creating tags from Genres, Keywords, Overview, Cast and Crew


```python
#Genres, keywords, cast and crew are of type list of dictionaries 
movies.genres[0]
```

```python
#Helper function to convert list of dictionaries into string
def get_tags(data):    
    #eval() converts str into list of dictionaries
    tagList = []
    for dict in eval(data):
        tagList.append(dict['name'].replace(" ","_"))     #Concatening space separated words together using '_'
    return " ".join(tagList)

def get_top3_actors(cast):
    actors = []
    for dict in eval(cast)[:3]:
        actors.append(dict['name'].replace(" ","_"))
    return " ".join(actors)

def get_director(crew):
    for dict in eval(crew):
        if dict['job'] == 'Director':
            return dict['name'].replace(" ","_")
```

```python
#Applying the transformations
movies.genres = movies.genres.apply(get_tags)
movies.keywords = movies.keywords.apply(get_tags)
movies.cast = movies.cast.apply(get_top3_actors)
movies.crew = movies.crew.apply(get_director)
movies.head()
```

```python
movies.info()
```

```python
#Replacing NA with empty string
movies.crew.fillna("",inplace=True)
```

```python
movies.info()
```

```python
#Concatenating all tags together
movies['tag']= movies.overview+" "+movies.genres+" "+movies.keywords+" "+movies.cast+" "+movies.crew
```

```python
movies.head()
```

```python
movies.tag[0]
```

## Text Preprocessing

```python
#Creating a new df with relevant columns only
movies_tag = movies[['id','title','tag']]
movies_tag.head()
```

```python
#Converting the tags into lowercase
movies_tag.tag = movies_tag.tag.apply(lambda x : x.lower())
movies_tag.head()
```

```python
#Removing Punctuation and special symbols (except '_' b/c we used it to concat words earlier)

punctuation = '`~!@#$%^&*()+-={}[]|\:;–"\'’,./<>?0123456789'
def remove_punctuation(text):
    for p in punctuation:
        if p in text:
            text = text.replace(p,"")
    return text
```

```python
movies_tag.tag = movies_tag.tag.apply(remove_punctuation)
movies_tag.head()
```

```python
#Stemming
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    words = []
    
    for word in text.split():
        words.append(ps.stem(word))
    return " ".join(words) 
```

```python
movies_tag.tag = movies_tag.tag.apply(stem)
movies_tag.head()
```

## Text Vectorization
**Bag of Words**:<br>
In Bag of Words, we will combine all the tags of all the movies and then count their frequencies. Top N words will be selected on the basis of their count. Stop words like 'in', 'and', 'it' etc. will not be considered. <br>

We will construct a matrix where the rows correspond to the movies and columns correspond to the Top N words (known as Document Term Matrix), which basically describes the frequency of tags corresponding to a particular movie. 

Eg. Consider<br>
M1 = "Fun world fun games"<br>
M2 = "Play games world enjoy " <br>

Document Term Matrix:<br>

        Fun Play World Games Enjoy
    M1   2  0    1     1     0
    M2   0  1    1     0     1

Here N = 5

Consider the N words as N dimensional vector. Vectors which are closer to each other are more similar. Here, we will use **Cosine Distance** (angle between them) to calculate the distance (hence similarity) between the vectors.<br>

**Euclidean Distance is not a reliable measure when dealing with High Dimensional Data.**

```python
#Importing the library to implement Text Vectorization
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(movies_tag['tag']).toarray()
vectors
```

```python
#To get the max features selected through CountVectorizer
cv.get_feature_names()[:10]
```

```python
#Calculating the cosine similarity (invervse of cosine distance)
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(vectors)
```

```python
#Entry (i,j) means similary of ith movie with jth movie
pd.DataFrame(similarity_matrix)
```

## Get Recommendations

```python
def recommend(movie):
    index = movies_tag[movies_tag['title']==movie].index.values[0] #Retrieve the index of the given movie
    top5_movies = sorted(enumerate(similarity_matrix[index]),reverse=True,key=lambda x : x[1])[1:6] #Find top 5 movies
    
    for x in top5_movies:
        print(movies_tag.title[x[0]])
```

```python
#Example
recommend('Batman')
```
